Variance Inflation Factor (VIF) method was applied to detect and drop highly collinear features. Highly collinear features were dropped because multicollinearity introduces several problems in statistical modeling and machine learning.

```python
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Encode target variable for numeric processing
df["Target"] = df["Target"].map({"Graduate": 1, "Dropout": 0})
```

```python
# Select numeric columns excluding target for VIF
numeric_df = df.select_dtypes(include=[np.number]).drop(columns=["Target"])
```

```python
# Compute VIF for each feature
vif_data = pd.DataFrame()
vif_data["Feature"] = numeric_df.columns
vif_data["VIF"] = [variance_inflation_factor(numeric_df.values, i) for i in range(numeric_df.shape[1])]
```

```python
vif_data
```

```python
# Drop features with VIF > 10
high_vif_features = vif_data[vif_data["VIF"] > 10]["Feature"].tolist()
df.drop(columns=high_vif_features, inplace=True)
```

```python
# Calculate fairness metrics: True Positive Rate (TPR) per group
# function to calculate TPR per group
def compute_tpr(group, group_name):
    grouped = df.groupby(group)
    result = pd.DataFrame()
    result["Group"] = grouped.groups.keys()
    result["Total"] = grouped["Target"].count().values
    result["Graduated"] = grouped["Target"].sum().values
    result["Dropout"] = result["Total"] - result["Graduated"]
    result["TPR (Graduation Rate)"] = result["Graduated"] / result["Total"]
    result["Feature"] = group_name
    return result[["Feature", "Group", "Total", "Graduated", "Dropout", "TPR (Graduation Rate)"]]
```

```python
# Metrics for Gender, Scholarship Holder, Displaced
gender_fairness = compute_tpr("Gender", "Gender")
scholar_fairness = compute_tpr("Scholarship_Holder", "Scholarship_Holder")
displaced_fairness = compute_tpr("Displaced", "Displaced")
```

```python
# Combine all
fairness_dashboard = pd.concat([gender_fairness, scholar_fairness, displaced_fairness], ignore_index=True)
fairness_dashboard.sort_values(by=["Feature", "TPR (Graduation Rate)"], ascending=[True, False])
```

```python
sample_img = x_train[0]  # Take a sample image
```

```python
plt.imshow(cv2.cvtColor(sample_img, cv2.IMREAD_GRAYSCALE))
plt.title("Sample Image")
plt.axis('off')
plt.show()
```

```python
# Normalize (0â€“1)
img_norm = sample_img / 255.0
```

```python
# Standardize (per-channel)
img_std = (img_norm - img_norm.mean()) / img_norm.std()
```

```python
# Grayscale
img_gray = cv2.cvtColor((img_std * 255).astype(np.uint8), cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
```

```python
# Resize (to 64x64)
img_resized = cv2.resize(img_gray, (64, 64), interpolation=cv2.INTER_LINEAR)
plt.imshow(img_resized)
plt.title("Resized Image")
plt.axis('off')
plt.show()
```

```python
# Noise Reduction (Gaussian blur)
img_denoised = cv2.GaussianBlur(img_resized, (5, 5), 0)
plt.imshow(img_denoised)
plt.title("Noisy Image")
plt.axis('off')
plt.show()
```

```python
# Enhancement (CLAHE - Adaptive Equalization)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
img_clahe = clahe.apply(img_denoised)
plt.imshow(img_clahe)
plt.title("Enhanced Image")
plt.axis('off')
plt.show()
```

```python
# Histogram Equalization
img_eq = cv2.equalizeHist(img_clahe)
plt.imshow(img_eq)
plt.title("Histogram Equalized Image")
plt.axis('off')
plt.show()
```

```python
# Edge Detection (Canny)
img_edges = cv2.Canny(img_eq, threshold1=50, threshold2=150)
plt.imshow(img_edges)
plt.title("Edge Detected Image")
plt.axis('off')
plt.show()
```

```python
# Segmentation (Simple K-means)
Z = img_eq.reshape((-1, 1)).astype(np.float32)
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
K = 3
_, labels, centers = cv2.kmeans(Z, K, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
segmented = centers[labels.flatten()].reshape(img_eq.shape).astype(np.uint8)
plt.imshow(segmented)
plt.title("Segmented Image")
plt.axis('off')
plt.show()
```

```python
x_data = x_train[:500]
y_data = y_train[:500].flatten()
```

```python
def preprocess_image(img):
    img_norm = img / 255.0
    img_std = (img_norm - img_norm.mean()) / img_norm.std()
    img_gray = cv2.cvtColor((img_std * 255).astype(np.uint8), cv2.COLOR_RGB2GRAY)
    img_resized = cv2.resize(img_gray, (64, 64), interpolation=cv2.INTER_LINEAR)
    img_denoised = cv2.GaussianBlur(img_resized, (5, 5), 0)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    img_clahe = clahe.apply(img_denoised)
    img_eq = cv2.equalizeHist(img_clahe)
    return img_eq
```

```python
# Apply preprocessing to all images
processed_images = np.array([preprocess_image(img) for img in x_data])
flat_images = processed_images.reshape(len(processed_images), -1)  # shape: (500, 4096)
```

```python
# Standardize and apply PCA
scaler = StandardScaler()
X_std = scaler.fit_transform(flat_images)
```

```python
pca = PCA(n_components=50)
X_pca = pca.fit_transform(X_std)
```

```python
# Plot cumulative variance
explained_variance = np.cumsum(pca.explained_variance_ratio_)
plt.figure(figsize=(8, 4))
sns.lineplot(x=range(1, 51), y=explained_variance)
plt.axhline(0.90, linestyle='--', color='red', label="90% Variance")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Variance Explained")
plt.title("PCA on Preprocessed CIFAR-10 Images")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 2D PCA Scatter Plot
pca2d = PCA(n_components=2)
X_pca_2d = pca2d.fit_transform(X_std)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca_2d[:, 0], y=X_pca_2d[:, 1], hue=y_data, palette="tab10", legend=False)
plt.title("2D PCA Projection of Preprocessed CIFAR-10 Images")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.tight_layout()
plt.show()
```
